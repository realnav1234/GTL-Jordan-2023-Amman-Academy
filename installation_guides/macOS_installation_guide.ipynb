### Part 2: Installing Python

First, let's check if Python is already installed on your computer. Go to Spotlight Search and type 'Terminal', which is the command terminal native to macOS. Open it when it pops up.



To make sure Python is not installed, type `python3` in the Terminal. This should give an error.

Note: if you do have Python installed, but it is not version 3.5 or later, reinstall Python regardless.

If Python is not installed, go to https://www.python.org/downloads/ to download Python. 

It should automatically recognize your operating system as macOS, and give you the option to install **Python 3.11.1**. If it does not, click the correct operating system and continue from there.



After downloading, accept all the licensing agreements and click 'Install'. Enter your password if prompted.

To make sure Python is installed, go back to Powershell, and type `python3`, and the version should say 3.11.1, or something close. To suspend the process, press **cmd + z**.

To make sure Package Installer for Python, or `pip`, is installed, type `pip3` into the terminal, and a slew of commands should show up.

### Part 4: Installing Packages

We will be using a variety of Python packages over the course of the next few weeks. We will use the Package Installer for Python `pip` for this. 

Go to Powershell, and run the following command to install `numpy`, a Python library used for math calculations.

- `pip3 install numpy`

If that worked, run the following command to install the other packages you would need for the course.

- `pip3 install torch torchvision torchaudio nltk matplotlib scipy pandas jupyter`

### Part 6: Installing Git

In order to install Git, we must install **Homebrew**, a package installer for Mac. Open Terminal, and copy and paste EXACTLY the following line into your terminal. This will run the Homebrew installation script.

`/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"`

*Note*: Press RETURN if prompted.

After this is complete, type `brew -v` to check if Homebrew has been installed.

To install Git, type `brew install git` into the terminal. 

After this is complete, run the command `git` in the terminal, and a slew of options should pop up indicating git has been installed.
